```python
src_folder = 'photos_src/'
dst_folder = 'photos_dst/'
src_files = ['faces1-11x11-704x704.png', 'faces2-14x10-896x640.png', 'faces3-13x13-832x832.png', 'faces4-8x8-512x512.png', 'faces5-19x7-1216x448.png', 'faces6-17x7-1088x448.png', 'faces7-11x11-704x704.png', 'faces8-12x13-768x832.png', 'faces9-14x13-896x832.png', 'faces10-12x12-768x768.png']
src_shapes = [(11,11), (14,10), (13,13), (8,8), (19,7), (17,7), (11,11), (12,13), (14,13), (12,12)]
```

```python
face_ind = 0
face_side_length = 64
```

```python
def extract_faces(src_file, rows, cols):

  global face_ind
  print(f"Source file: {src_file}\n")

  # open faces set
  print("Original image")
  faces = PIL.Image.open(src_folder+src_file)
  print(f"Image size: {faces.size}")
  print(f"Image mode: {faces.mode}")
  faces_array = np.asarray(faces)
  print(f"Data type: {type(faces_array)}")
  print(f"Data shape: {faces_array.shape}")
  print("Preview:")
  preview = faces.copy()
  preview.thumbnail((100,100))
  preview.show()

  # change to greyscale
  print("\nGrayscale image")
  grayscale_faces = faces.convert('L')
  print(f"Image size: {grayscale_faces.size}")
  print(f"Image mode: {grayscale_faces.mode}")
  grayscale_faces_array = np.asarray(grayscale_faces)
  print(f"Data type: {type(grayscale_faces_array)}")
  print(f"Data shape: {grayscale_faces_array.shape}")
  print("Preview:")
  preview = grayscale_faces.copy()
  preview.thumbnail((100,100))
  preview.show()

  # cut faces, remove background and save to separate files
  print("\n")
  for row in range(rows):
    for col in range(cols):
      face_ind += 1
      print(f"\rSaving face nr {face_ind}", end="")
      face_array = grayscale_faces_array[row*face_side_length : (row+1)*face_side_length, col*face_side_length : (col+1)*face_side_length]
      face = PIL.Image.fromarray(face_array)
      face_cleaned = remove(face)
      face_cleaned.save(dst_folder+'face'+str(face_ind)+'.png')

  print(f"\nFaces saved: {rows*cols}, up to index: {face_ind}")
  print("Last face from file:")
  face_cleaned.show()
  print("===========================================\n")
```

```python
extract_faces(src_files[0], 11, 11)
```
